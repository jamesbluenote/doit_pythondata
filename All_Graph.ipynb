# 2. 막대 그래프 (평균값 등)

## @ [seaborn] sns.barplot()

### - 평균표/빈도표 등을 먼저 만들어야 함

```python
df_mpg_drv = df_mpg.groupby('drv', as_index = False) \
                   .agg(mean_hwy = ('hwy', 'mean')) \
                   .sort_values('mean_hwy', ascending = False)
df_mpg_drv
```

```python
sns.barplot(data = df_mpg_drv, x = 'drv', y = 'mean_hwy');
```

### - 막대 크기순으로 정렬

```python
df_mpg_drv = df_mpg_drv.sort_values('mean_hwy', ascending = False)
sns.barplot(data = df_mpg_drv, x = 'drv', y = 'mean_hwy');
```

# 3. 누적 비율 막대 그래프

## @ [plt] df.plot.barh(stacked=True)

### - 빈도표 만들기

```python
df_mpg = mpg.groupby('category', as_index = False) \
             ['drv'] \
            .value_counts(normalize = True)
df_mpg = df_mpg.assign(proportion = df_mpg['proportion']*100).round(1)
df_mpg
```

### - pivot

```python
mpg_pivot = df_mpg.pivot(index = 'category', 
                         columns = 'drv',
                         values = 'proportion')
mpg_pivot
```

```python
plt.rcParams.update({'figure.dpi' : '100',
                     'figure.figsize' : (6, 4)})
```

```python
mpg_pivot.plot.barh(stacked = True);
```

### - 막대 정렬

```python
mpg_pivot_reorder = mpg_pivot.sort_values('f')[['f', 'r', '4']]
```

```python
mpg_pivot_reorder.plot.barh(stacked = True);
```

# 7. 선 그래프

## @ [seaborn] sns.lineplot()

```python
economics_raw = pd.read_csv('economics.csv')
economics = economics_raw.copy()
economics.head()
```

```python
sns.lineplot(data = economics, x = 'date', y = 'unemploy');
```

### - x축에 연도만 표시

#### (1) 날짜 -> datetime 타입으로 변경

```python
economics['date2'] = pd.to_datetime(economics['date'])
economics.info()
```

```python
economics[['date', 'date2']].head()
```

#### (2) 연도만 추출

```python
economics['year'] = economics['date2'].dt.year
economics.head()
```

#### (3) 선 그래프 그리기

```python
sns.lineplot(data = economics, x = 'year', y = 'unemploy');
```

### - 신뢰구간(Confidence Interval) 제거

```python
sns.lineplot(data = economics, x = 'year', y = 'unemploy', ci = None);
```
