# ------------------------------------------------------

# 1. 빈도표

## @ [pandas] df.plot.bar()

### - 빈도 순서(내림차순)로 막대 정렬

```python
count = df_mpg['manufacturer'].value_counts()
```

```python
count.plot.bar(rot = 45);
```

### - 빈도 순서(오름차순)로 막대 정렬

```python
count_desc = df_mpg['manufacturer'].value_counts().sort_values(ascending = True)
```

```python
count_desc.plot.bar(rot = 45);
```

### - 인덱스의 알파벳 순서(오름차순)로 막대 정렬

```python
count_sorted = df_mpg['manufacturer'].value_counts().sort_index()
```

```python
count_sorted.plot.bar(rot = 45);
```

### - 인덱스의 알파벳 순서(내림차순)로 막대 정렬

```python
count_sorted_desc = df_mpg['manufacturer'].value_counts().sort_index(ascending = False)
```

```python
count_sorted_desc.plot.bar(rot = 45);
```

## @ [seaborn] sns.countplot()

```python
# 알파벳 순서로 막대 정렬
sns.countplot(data = df_mpg, y = 'manufacturer');
```
