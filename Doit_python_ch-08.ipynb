## 08-3. 막대 그래프 - 집단 간 차이 표현하기

### 평균 막대 그래프 만들기
#### 1. 집단별 평균표 만들기

```python
df_mpg = mpg.groupby('drv', as_index = False).agg(mean_hwy = ('hwy', 'mean'))
df_mpg
```

#### 2. 그래프 만들기

```python
sns.barplot(data = df_mpg, x='drv', y='mean_hwy');
```

#### 3. 크기순으로 정렬하기

```python
df_mpg = df_mpg.sort_values('mean_hwy', ascending = False)
```

```python
sns.barplot(data = df_mpg, x='drv', y='mean_hwy');
```

### 빈도 막대 그래프 만들기
#### 1. 집단별 빈도표 만들기

```python
df_mpg = mpg.groupby('drv', as_index = False).agg(n = ('drv', 'count'))
df_mpg
```

#### 2. 그래프 만들기

```python
sns.barplot(data = df_mpg, x='drv', y='n');
```

#### @ sns.countplot() 을 이용하면 df.groupby().agg() 없이 바로 barplot 그릴 수 있음

```python
sns.countplot(data = mpg, x='drv');
```

#### 막대 정렬하기

```python
# 순서 내가 정하기
sns.countplot(data = mpg, x='drv', order = ['4', 'f', 'r']);
```

```python
# 높은 빈도순으로 출력
sns.countplot(data = mpg, x='drv', order = mpg['drv'].value_counts().index);
```

### p.211 - 혼자서 해보기

```python
# Q1
mpg_raw = pd.read_csv('mpg.csv')
mpg = mpg_raw.copy()
mpg.head()
```

```python
mpg_suv = mpg.query('category == "suv"') \
             .groupby('manufacturer', as_index = False) \
             .agg(mean_cty = ('cty', 'mean')) \
             .sort_values('mean_cty', ascending = False) \
             .head()
mpg_suv
```

```python
# Q2-2
# sns.barplot()
mpg_cat = mpg.groupby('category', as_index = False).agg(n = ('category', 'count')).sort_values('n', ascending = False)
mpg_cat
```

```python
sns.barplot(data=mpg_cat, x='category', y='n');
```

## 08-4. 선 그래프 - 시간에 따라 달라지는 데이터 표현하기
### 시계열 그래프 만들기

```python
economics_raw = pd.read_csv('economics.csv')
economics = economics_raw.copy()
economics.head()
```

```python
economics.shape
```

```python
economics.info()
```

```python
economics.describe()
```

```python
sns.lineplot(data=economics, x='date', y='unemploy');
```

#### x축에 연도 표시하기
##### (1) 날짜 시간 타입 변수 만들기

```python
economics.assign(date2 = lambda x: pd.to_datetime(x['date']))
```

```python
economics['date2'] = pd.to_datetime(economics['date'])
economics.info()
```

```python
economics[['date', 'date2']].head()
```

```python
economics['date2'].dt.year
```

```python
economics['date2'].dt.month
```

```python
economics['date2'].dt.day
```

##### (2) 연도 변수 만들기

```python
economics['year'] = economics['date2'].dt.year
economics.head()
```

##### (3) x축에 연도 표시하기

```python
sns.lineplot(data=economics, x='year', y='unemploy');
```

```python
# 신뢰구간(confidence interval) 미표시
sns.lineplot(data=economics, x='year', y='unemploy', ci=None);
```
